```python
import pandas as pd
import numpy as np
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
import pickle
```

```python
#Read the stock data
colnames=['Date', 'CompanyName', 'ClosingPrice', 'ChangeInStockPrice']
stockData = pd.read_csv("/Users/Anshu/MLCS_Project_2018-master/stockDataFinalEnhanced500.csv",names=colnames, header=None)
print(len(stockData))

stockData
```

```python
#stockData = stockData.rename(columns={'0': 'Date', '1': 'CompanyName','3':'ChangeInStockPrice'})
stockData['Date'] = pd.to_datetime(stockData['Date'])

#Number of data points
print(len(stockData))

#Find if any column has missing values
#print(stockData.columns[stockData.isnull().any()])

#stockData = stockData[]


print(stockData.dtypes)
```

```python
#Read the train case data
caseDataFile_train = open("/Users/Anshu/MLCS_Project_2018-master/traindocvector.pkl","rb")
caseData_train = pickle.load(caseDataFile_train)

print(len(caseData_train))
print(caseData_train)
```

```python
#Read the test case data
caseDataFile_test = open("/Users/Anshu/MLCS_Project_2018-master/testdocvector.pkl","rb")
caseData_test = pickle.load(caseDataFile_test)

print(caseData_test)
len(caseData_test.head(10))
```

```python
#Make the date format of all the datasets same
caseData_train['Date'] =  pd.to_datetime(caseData_train['Date'])
caseData_test['Date'] =  pd.to_datetime(caseData_test['Date'])
```

```python
#Join stock and train case data
joined_train_data = pd.merge(caseData_train,stockData,how='inner',on='Date')

#joined_train_data = joined_train_data.head(11000)

#Join stock and test case data
joined_test_data = pd.merge(caseData_test,stockData,how='inner',on='Date')

#joined_test_data = joined_train_data.head(1000)
```
