```python
import matplotlib.pyplot as plt
import numpy as np

```

# Visualization of Entire Data

```python
actual_all = np.loadtxt('actual.txt', dtype=float)
predicted_all = np.loadtxt('prediction.txt', dtype=float)

actual=actual_all[0:400]
predicted=predicted_all[0:400]
x = range(len(actual))

fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.scatter(x, actual, s=7, c='b', marker="s", label='actual change in stock')
ax1.scatter(x,predicted, s=7, c='r', marker="o", label='predicted change in stock')

#ax1.plot(x, actual, c='b', marker="s", label='actual change in stock')
#ax1.plot(x,predicted, c='r', marker="o", label='predicted change in stock')
plt.legend(loc='upper left');
plt.ylim(-3,3)
plt.show()

fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.scatter(x, actual, s=7, c='b', marker="s", label='actual change in stock')
#ax1.plot(x, actual, c='b', marker="s", label='actual change in stock')
plt.legend(loc='upper left');
plt.ylim(-3,3)
plt.show()

fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.scatter(x,predicted, s=7, c='r', marker="o", label='predicted change in stock')
plt.legend(loc='upper left')
plt.ylim(-3,3)
plt.show()
```

# Actual vs predicted

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.scatter(predicted, actual, s=7, c='b', marker="s", label='actual vs predicted')
#ax1.scatter(x,predicted, s=7, c='r', marker="o", label='predicted change in stock')

#ax1.plot(x, actual, c='b', marker="s", label='actual change in stock')
#ax1.plot(x,predicted, c='r', marker="o", label='predicted change in stock')
plt.legend(loc='upper left');
plt.xlabel('predicted')
plt.ylabel('actual')
plt.show()
```
